## Problem 1

In our first example we are looking at the temperature reading (meta-data) associated with an experiment. For the experiment to work reliably, the temperature should be at around 12 Kelvin, and if we look at the data it is mostly consistent with 12 Kelvin to within the 0.4 degree precision of the thermometry and the thermal control system (standard deviation).

Generate 100k data points based on a Gaussian distribution with mean 12 and std. deviation 0.4, representing what the temperature should be if the experiment behaves well. Then add the anomalous readings to create a representation of the flawed data set we got from the experiment.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from collections import Counter
```

```python
anom = [10., 10.3, 2.1, 0., 0., 15.6, 22.3, 12.7]
dat = stats.norm.rvs(loc=12, scale=0.4, size=100000)
anomdat = np.append(dat, anom)

plt.rcParams["figure.figsize"] = (15,10)
plt.tick_params(labelsize = 24)

plt.hist(anomdat, 200, density=True)
plt.yscale('log')
plt.ylabel('probability')
plt.xlabel('Temperature (K)')
plt.show()
```

The above graph is a log-scale histogram of our contaminated data. We want to throw out the anomalous data from when the thermal control system was misbehaving or not working. The statistical question we seek to answer is "Which of these temperature data points are within 5σ of the background distribution?" (or equivalently, "Which data points are malfunctions or outliers with greater than 5σ certainty, which we will remove from our data?") and the answer is the data which we want to keep, as it is *most* likely real data that isn't contaminated with the anomalous readings.

In math,

Let $D$ be our clean data and $X$ be our plotted data.

$d = x$ $\forall$ $x$ where $p(x)  < 5σ$

```python
X = np.sort(anomdat)
D = np.array([])
garb = np.array([])

fivesig = stats.norm.sf(5)
maxTemp = stats.norm.isf(fivesig, loc=12, scale=0.4)
minTemp = stats.norm.isf(1-fivesig, loc=12, scale=0.4)

D = X[np.logical_and(X > minTemp, X < maxTemp)]
garb = X[np.logical_or(X < minTemp, X > maxTemp)]
print("Anomalous data added :", anom)
print("Number of fake data points :", len(anom))
print("Thrown out values :", garb)
print("Number of thrown values :", len(garb))
print("Size of D:", len(D))
```

So now we construct a 'Truth Table,'  

```python
truepos = sum((Counter(dat) & Counter(D)).values())
falsepos = sum((Counter(D) & Counter(anom)).values())
trueneg = sum((Counter(dat) & Counter(garb)).values())
falseneg = sum((Counter(garb) & Counter(anom)).values())

table = pd.DataFrame(data=[[truepos,falsepos],[trueneg,falseneg]],
                     index=["Pos", "Neg"],
                     columns=["True", "False"])
table
```

If we were to select a significance threshold less than 5 sigma (including all data up to threshold <5σ), there would be less false positives, more true negatives and more false negatives. If we we included all data at greater than 5 sigma significance, there would be more false positives, less true negatives, and less false negatives.

For example, lets put the significance threshold at 3σ

```python
fivesig = stats.norm.sf(3)
maxTemp = stats.norm.isf(fivesig, loc=12, scale=0.4)
minTemp = stats.norm.isf(1-fivesig, loc=12, scale=0.4)

D = X[np.logical_and(X > minTemp, X < maxTemp)]
garb = X[np.logical_or(X < minTemp, X > maxTemp)]
print("Number of fake data points :", len(anom))
print("Number of thrown values :", len(garb))
print("Size of D:", len(D))
```

```python
truepos = sum((Counter(dat) & Counter(D)).values())
falsepos = sum((Counter(D) & Counter(anom)).values())
trueneg = sum((Counter(dat) & Counter(garb)).values())
falseneg = sum((Counter(garb) & Counter(anom)).values())

table = pd.DataFrame(data=[[truepos,falsepos],[trueneg,falseneg]],
                     index=["Pos", "Neg"],
                     columns=["True", "False"])
table
```

As stated above, the number of true positives was decreased by one and the number of false negatives increased by one, at the cost of 281 erroneously thrown out data points.

For the significance threshold at 7σ

```python
fivesig = stats.norm.sf(7)
maxTemp = stats.norm.isf(fivesig, loc=12, scale=0.4)
minTemp = stats.norm.isf(1-fivesig, loc=12, scale=0.4)

D = X[np.logical_and(X > minTemp, X < maxTemp)]
garb = X[np.logical_or(X < minTemp, X > maxTemp)]
print("Number of fake data points :", len(anom))
print("Number of thrown values :", len(garb))
print("Size of D:", len(D))
```

```python
truepos = sum((Counter(dat) & Counter(D)).values())
falsepos = sum((Counter(D) & Counter(anom)).values())
trueneg = sum((Counter(dat) & Counter(garb)).values())
falseneg = sum((Counter(garb) & Counter(anom)).values())

table = pd.DataFrame(data=[[truepos,falsepos],[trueneg,falseneg]],
                     index=["Pos", "Neg"],
                     columns=["True", "False"])
table
```

Again, when increasing the range of acceptable data (requires 7σ certainty that a data point is an anomaly before it is removed) we increase the number of false positives in return for greatly reducing the number of erroneous omissions (true negatives) and also correct omissions (false negatives).

There are two mistakes of comission in the 5σ threshold (the values being 12.7 and 10.3), and they are *technically* avoidable; however reducing the statstical threshold to declare the value 12.7 an anomaly would also cause over ~10% of the (mostly real!) data to be thrown out. The threshold required would be about 1.5σ, and 13000 values are erroneously thrown out to remove all mistakes of commission.

## Problem 2

If we look at the alignment of stars on subsequent images, they don't perfectly align due to atmospheric and instrumental effects (even ignoring proper motion). The resulting distribution is two dimensional, and for this lab let's assume it is a 2D Gaussian with 1 arcsecond RMS.

```python
#a = np.concatenate(np.array(stats.norm.rvs( scale = 1, size = 100000)), np.array(stats.norm.rvs( scale = 1, size = 100000)), axis = 1)
a = np.vstack((stats.norm.rvs( scale = 1, size = 100000), stats.norm.rvs( scale = 1, size = 100000)))
a.shape
fig, ax = plt.subplots(1, 1)
h = ax.hist2d(a[0,:],a[1,:],bins=100, density=True);
ax.set_aspect('equal', 'box')
plt.xlim([-3 , 3])
plt.ylim([-3 , 3])
plt.title("2D Histogram of positional uncertainty", fontsize = 24)
plt.ylabel("$\Delta$y arcseconds", fontsize = 18)
plt.xlabel("$\Delta$x arcseconds", fontsize = 18)
plt.colorbar(h[3], ax=ax)
plt.show()
```

A potential asteroid would have true movement between the images; that is, there would be some data point with ($\Delta x$,$\Delta y$) displacement. The real distance is given by $$D = \sqrt{\Delta x^2 + \Delta y^2}$$

Given that both x and y are gaussian distributed, the new distribution D is made by the square root of the sum of both squared gaussian distributions. The distribution D then is a chi squared distribution with a degree of freedom of two, or a Rayleigh distribution. Because $\Delta x$ and $\Delta y$ have a standard deviation of one and mean of zero, the Rayleigh distribution has the default mean of $\sqrt{\frac{\pi}{2}}$ and variance of $\frac{4-\pi}{2}$.

The question we want to answer is "What distance would an asteroid have to travel (in arcseconds) to have 5σ certainty that it wasn't movement created by random noise from the atmospheric and instrumental effects?"

In math, that means find $\tilde{D}$, the distance required to create a signal of 5 sigma certainty or greater. 
$D \in $ Chi Square Distribution  
$5\sigma = 2.866e-07$  
$\tilde{D} =$ rayleigh.ppf($1-5\sigma$)  

```python
fivesig = stats.norm.sf(5)
print("Distance in arcseconds : ", stats.rayleigh.ppf(1-fivesig))
```

## Problem 3
##### Background
Cosmic rays are charged particles—usually protons or electrons but can include atomic nuclei such a alpha particles (helium) or iron. Because of their charge cosmic rays spiral in the magnetic field of the galaxy. From the perspective of the Earth they appear to be coming uniformly from all directions like a high energy gas, and the direction the cosmic ray is travelling when it reaches the Earth tells us nothing about where it came from because we don't know what tortured path it has taken through the galaxy to reach us. However, at trillion electron volt energies and above, the spiral loops are fairly big and the sun and the moon will block cosmic rays. This means the sun and the moon appear as holes in the cosmic ray sky (cosmic rays from that direction are absorbed).

Assume a moon sized patch on the sky we normally have a cosmic ray rate of 1 cosmic ray per minute (arrivals are random in time). If we can observe where the moon is for 8 hours per night (not too close to the horizon) and we observe for 15 days and see 6800 cosmic rays, what is the signficance of our moon shadow detection?

##### Analysis
In other words, what is the probability that we see 6800 cosmic rays over 15 days, coming from a background which produces 1 cosmic ray per minute? What is the significance (in sigma) of the signal, if there is one?

The background is discrete, given that cosmic rays come in integer amounts, therefore the probability mass function of a cosmic ray arriving for a given minute is most likely a Poisson distribution with the given average of one per minute. Summed over 15 days, 8 hours per day, comes to 7,200 samples, it will approach a Gaussian distribution by the central limit theorem. The trials factor is not particularly relevant as the source we are looking for is persistent, and one particularly unusual event in a sample won't be mistaken for a signal in this case; the relevant number is the sum of all samples.

Let $P$ be the Poisson distribution with a mean of $1$. Find the probability of a signal of strength $Y=6800$ over $N=7200$ samples, given a probability mass function $f(p)$.


```python
plt.tick_params(labelsize = 24)
mu = 1
xbase = np.arange(stats.poisson.ppf(0.00001, mu), 
              stats.poisson.ppf(0.99999, mu))
pmf = stats.poisson.pmf(xbase, mu)
plt.plot(xbase, pmf, 'bo', ms=8, label='poisson pmf')
plt.vlines(xbase, 0, pmf, colors='b', lw=5, alpha=0.5)
plt.show()
```

This graph represents $f(p)$, the probability mass function of cosmic rays entering from the moon's shadow over one minute. 

Next, sum the number of cosmic rays over one hour

```python
n = 60
sumpmf = pmf
for i in range(n-1):
    sumpmf = np.convolve(sumpmf, pmf)
    
x = np.arange(xbase[0]*n, xbase[-1]*n+1)
plt.plot(x, sumpmf, 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, sumpmf, colors='b', lw=5, alpha=0.5)
plt.xlim([30,100])
plt.show()
```

This graph represents the sum of the background over one hour, and looks exactly the same as a Poisson distribution with a mean of 60 (which is to be expected given 60 minutes with an average of 1 cosmic ray per minute). Applying this to create a probability distribution for the number of cosmic rays detected over all 15 days,

```python
mu = 15*8*60
x = np.arange(stats.poisson.ppf(0.00001, mu), 
              stats.poisson.ppf(0.99999, mu))
pmf = stats.poisson.pmf(x, mu)
plt.plot(x, pmf, 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, pmf, colors='b', lw=5, alpha=0.5)
plt.show()
```

In the stated question the signal is $Y=6800$. The above graph represents our probability mass function $\tilde{f}(p)$, the probability for a given number of cosmic rays from the background over 15 days. Let $\tilde{F}(p)$ be the cumulative distribution function for this distribution. Find $\tilde{F}(6800)$ 

In math,
$$\tilde{F}(6800) = \int_0^{6800} f(p)dp = \sum_0^{6800} f(p)$$

Then calculate the sigma significance, stats.norm.ppf$(1-\tilde{F}(6800))$

```python
prob = stats.poisson.cdf(6800, mu)
sig = stats.norm.ppf(1-prob)
print("Probability of 6800 rays over 15 days : ", prob)
print("Sigma significance of discovery : ", sig)
print(1/prob)
```

In answer to the original question, the probability of 6,800 cosmic rays coming from the background is approximately one in one million. With a sigma significance of 4.75, there is almost enough evidence to claim a discovery that the moon is absorbing the cosmic rays detected by the instruments.
